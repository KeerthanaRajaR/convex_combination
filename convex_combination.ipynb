```python
import numpy as np
import matplotlib.pyplot as plt

p1 =np.array([2,6])
p2 =np.array([5,8])

def linspace(start, stop, num):
    return [start + (stop - start) * i / (num - 1) for i in range(num)]

lambdas = linspace(0, 1, 10)
print(lambdas,"\n \r")
points =[l*p1 +(1-l)*p2 for l in lambdas]

x_vals ,y_vals =zip(*points)
plt.plot([p1[0], p2[0]], [p1[1], p2[1]], 'b--', label='Line between p1 and p2')
plt.scatter(x_vals, y_vals, color='r', label='Generated Points')
plt.scatter(*p1, color='y', label='p1')
plt.scatter(*p2, color='y', label='p2')
plt.legend()
plt.show()

```
