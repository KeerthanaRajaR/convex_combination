```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
lambda1 = np.linspace(0,1,20)
lambda2 = 1-lambda1
print("lambda1:\n \n",lambda1)
print("\nlambda2: \n \n",lambda2)
```

```python
p1=np.array([2,6])
p2=np.array([4,7])
```

```python
plt.scatter(lambda1*p1[0] + lambda2*p2[0] ,lambda1*p1[1] + lambda2*p2[1] , c='r')
plt.show()
```
